```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings

warnings.filterwarnings('ignore')

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'NanumGothic'
```

```python
Bike_Stop = pd.read_excel('./raw data/대여소/서울특별시 공공자전거 대여소 정보(19.12.9).xlsx')
Bike_Stop_ID = Bike_Stop.groupby('대여소_구')['대여소ID'].unique().loc['동대문구']
#Bike_Stop_ID = np.append(Bike_Stop_ID, 664)
Bike_Stop_ID = Bike_Stop_ID.astype(np.int64)
Bike_Stop_ID
```

### 소요시간의 평균 값 구하기

```python
import datetime
import time
import pickle
import numpy as np 
with open('./time/Bike_Stop_Time.pickle', 'rb') as fr:
    Time = pickle.load(fr)
```

```python
temp2 = {}
for k, l in zip(Bike_Stop_ID, range(51)):
    #temp1[str(k)] = min(list(user_loaded[str(k)].values()).remove(0))
    #print(del list(user_loaded[str(k)].values())[l])
    t = list(Time[str(k)].values())
    del t[l]
    t = np.array(t)
    #print(t)
    #print(min(t))
    temp2[str(k)] = np.mean(t)
```

```python
temp2
```

```python
temp3 = []
for k in temp2.values():
    temp3 += [k]
    
avg_time = np.mean(np.array(temp3))
avg_time
```
